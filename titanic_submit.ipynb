```python
import pandas as pd 
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df.Survived.value_counts()
```

```python
import seaborn as sns
```

```python
sns.set(style="whitegrid")

g = sns.catplot(x="Pclass", y="Survived", data=df,
                height=6, kind="bar", palette="muted")
g.despine(left=True)
```

### The plot shows the relation between survival probability and cabin class. It can be observed that survival probability decreases with cabin class downgrades.

```python
sns.catplot(x="Sex", y="Survived", hue="Pclass", kind="bar", data=df);
```

### The plot shows the relation between survival result and cabin class/sex. It can be observed that survival probability decreases with cabin class downgrades, and the condition is the same for both male and female.

```python
sns.lmplot("Fare", "Survived", df, logistic=True, truncate=True);
```

### The plot shows the relation between survival result and fare. It can be observed that survival probability increases with fare increases.
